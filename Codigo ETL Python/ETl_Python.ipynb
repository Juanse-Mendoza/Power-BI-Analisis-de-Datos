```python
import pandas as pd

# Define la ruta del archivo de Excel.
# Reemplaza 'nombre_de_tu_archivo.xlsx' con el nombre de tu archivo real.
ruta_archivo = '/content/Dataset con varios errores 222.xlsx'

# Carga el archivo de Excel en un DataFrame de pandas.
df = pd.read_excel(ruta_archivo)

# Muestra las primeras 5 filas del DataFrame para verificar que se cargó correctamente.
print(df.head())
```

```python

# Remplaza todos los valores nulos (NaN) en la columna 'Nombre_Completo' por 'sin_nombre'
df['Nombre_Completo'] = df['Nombre_Completo'].fillna('sin_nombre')

# Muestra el DataFrame modificado para verificar el cambio
print(df)
```

```python
# Reemplaza los valores de texto por sus equivalentes numéricos
df['Edad'] = df['Edad'].replace({'Veinte y dos': 22,
                                 'Treinta y seis': 36})

# Muestra el DataFrame modificado
print(df)
```

```python
# Reemplaza la cadena de texto 'null' por 'Sin_ciudad'
df['Ciudad'] = df['Ciudad'].replace('null', 'sin_ciudad')

# Reemplaza los valores nulos (NaN) por 'Sin_ciudad'
df['Ciudad'] = df['Ciudad'].fillna('sin_ciudad')

# Muestra el DataFrame modificado
print(df)
```

```python
# Reemplaza el símbolo de dólar ($) y las comas con una cadena vacía
df['Total_Gastado_USD'] = df['Total_Gastado_USD'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False)

# Convierte la columna a tipo de dato numérico (float)
df['Total_Gastado_USD'] = pd.to_numeric(df['Total_Gastado_USD'])

# Muestra el DataFrame modificado
print(df)
```

```python
# Remplaza todos los valores nulos (NaN) en la columna 'Producto_Favorito' por 'sin_producto'
df['Producto_Favorito'] = df['Producto_Favorito'].fillna('sin_producto')

# Muestra el DataFrame modificado para verificar el cambio
print(df)
```

```python
# Define la ruta y el nombre del nuevo archivo de Excel
ruta_archivo_final = 'mi_dataframe_limpio.xlsx'

# Exporta el DataFrame a un archivo de Excel
df.to_excel(ruta_archivo_final, index=False)

print(f"DataFrame exportado exitosamente a '{ruta_archivo_final}'")
```
