```python
import pandas as pd

# Define la ruta del archivo de CSV.
# Reemplaza 'nombre_de_tu_archivo.csv' con el nombre de tu archivo real.
ruta_archivo = '/content/dataset_streaming_500k_errores.csv'

# Carga el archivo de Excel en un DataFrame de pandas.
df = pd.read_csv(ruta_archivo)

# Muestra las primeras 5 filas del DataFrame para verificar que se cargó correctamente.
print(df.head())
```

```python
import numpy as np

# Intentar convertir la columna 'Edad' a tipo numérico. Los errores se convertirán en NaN.
df_cleaned['Edad'] = pd.to_numeric(df_cleaned['Edad'], errors='coerce')

# Rellenar los valores nulos (incluyendo aquellos que eran texto y no se pudieron convertir) con "no_edad"
df_cleaned['Edad'] = df_cleaned['Edad'].fillna('no_edad')

# Convertir los valores numéricos restantes a entero (los "no_edad" seguirán siendo cadenas)
# Primero, identificamos las filas que NO son 'no_edad'
numeric_ages_mask = df_cleaned['Edad'] != 'no_edad'

# Luego, convertimos solo esas filas a entero
df_cleaned.loc[numeric_ages_mask, 'Edad'] = df_cleaned.loc[numeric_ages_mask, 'Edad'].astype(int)


# Mostrar información sobre la columna 'Edad' y las primeras filas para verificar
print("\nInformación de la columna 'Edad' después de limpiar:")
print(df_cleaned['Edad'].value_counts(dropna=False).head()) # Mostrar la frecuencia de los primeros valores (incluido 'no_edad')
print("\nPrimeras filas del DataFrame después de limpiar 'Edad':")
print(df_cleaned.head())
```

```python
# Eliminar filas duplicadas basadas en la columna 'ID_Encuesta'
df_cleaned = df.drop_duplicates(subset=['ID_Encuesta'])

# Restablecer el índice para tener una numeración secuencial
df_cleaned = df_cleaned.reset_index(drop=True)

# Mostrar las primeras filas del DataFrame limpio para verificar
print(df_cleaned.head())
```

```python
# Verificar si hay valores nulos en la columna 'Nombre_Usuario'
print("Valores nulos antes de limpiar:", df_cleaned['Nombre_Usuario'].isnull().sum())

# Eliminar filas con valores nulos en 'Nombre_Usuario'
df_cleaned.dropna(subset=['Nombre_Usuario'], inplace=True)

# Estandarizar el formato a minúsculas
df_cleaned['Nombre_Usuario'] = df_cleaned['Nombre_Usuario'].str.lower()

# Verificar si hay nombres de usuario duplicados
print("Nombres de usuario duplicados antes de eliminar:", df_cleaned['Nombre_Usuario'].duplicated().sum())

# Eliminar filas con nombres de usuario duplicados (manteniendo la primera aparición)
df_cleaned.drop_duplicates(subset=['Nombre_Usuario'], keep='first', inplace=True)

# Restablecer el índice después de eliminar filas
df_cleaned.reset_index(drop=True, inplace=True)

# Mostrar las primeras filas del DataFrame limpio y el número de filas restante
print("\nDataFrame después de limpiar 'Nombre_Usuario':")
print(df_cleaned.head())
print("\nNúmero de filas después de limpiar 'Nombre_Usuario':", len(df_cleaned))
```

```python
# Estandarizar la columna 'País' a formato título
df_cleaned['País'] = df_cleaned['País'].str.title()

# Identificar las columnas desde 'País' hacia la derecha
columns_to_fill = df_cleaned.columns[df_cleaned.columns.get_loc('País'):]

# Rellenar valores nulos o vacíos con 'no_dato' en las columnas seleccionadas
df_cleaned[columns_to_fill] = df_cleaned[columns_to_fill].fillna('no_dato')

# Mostrar las primeras filas del DataFrame para verificar los cambios
print("DataFrame después de limpiar 'País' y rellenar nulos:")
print(df_cleaned.head())
```

```python
# Limpiar la columna 'Minutos_Vistos_Mes' eliminando comas y puntos que no sean decimales
# Primero, convertimos a string para asegurar que el método .str.replace funcione
df_cleaned['Minutos_Vistos_Mes'] = df_cleaned['Minutos_Vistos_Mes'].astype(str).str.replace('[^0-9.,]', '', regex=True)
df_cleaned['Minutos_Vistos_Mes'] = df_cleaned['Minutos_Vistos_Mes'].str.replace(',', '.') # Reemplazar comas por puntos decimales

# Intentar convertir la columna a tipo numérico (float)
df_cleaned['Minutos_Vistos_Mes'] = pd.to_numeric(df_cleaned['Minutos_Vistos_Mes'], errors='coerce')

# Rellenar los valores que no se pudieron convertir (NaN) con 0 o algún otro valor por defecto si es necesario
# En este caso, si no se pudo convertir, asumimos que el valor es 0 minutos
df_cleaned['Minutos_Vistos_Mes'] = df_cleaned['Minutos_Vistos_Mes'].fillna(0)

# Redondear a un máximo de dos decimales
df_cleaned['Minutos_Vistos_Mes'] = df_cleaned['Minutos_Vistos_Mes'].round(2)

# Mostrar información sobre la columna y las primeras filas para verificar
print("\nInformación de la columna 'Minutos_Vistos_Mes' después de limpiar:")
print(df_cleaned['Minutos_Vistos_Mes'].head())
print("\nTipo de dato de la columna 'Minutos_Vistos_Mes':", df_cleaned['Minutos_Vistos_Mes'].dtype)
print("\nPrimeras filas del DataFrame después de limpiar 'Minutos_Vistos_Mes':")
print(df_cleaned.head())
```
