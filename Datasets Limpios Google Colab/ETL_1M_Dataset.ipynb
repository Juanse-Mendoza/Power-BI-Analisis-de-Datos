```python
import pandas as pd

# Define la ruta del archivo de CSV.
# Reemplaza 'nombre_de_tu_archivo.csv' con el nombre de tu archivo real.
ruta_archivo = '/content/dataset_caotico_1m.csv'

# Carga el archivo de Excel en un DataFrame de pandas.
df = pd.read_csv(ruta_archivo)

# Muestra las primeras 5 filas del DataFrame para verificar que se cargó correctamente.
print(df.head())
```

```python
# Reemplazar valores nulos o vacíos con 'no_dato'
df.fillna('no_dato', inplace=True)

# Mostrar las primeras filas del DataFrame para verificar los cambios
print(df.head())
```

# Task
Clean the 'País' column in the dataframe by replacing inconsistent values (abbreviations, special characters, and misspellings) with their correct full names.

## Identificar inconsistencias

### Subtask:
Mostrar los valores únicos en la columna 'País' para identificar las inconsistencias.


**Reasoning**:
Access the 'País' column, get the unique values, and print them to identify inconsistencies as requested by the subtask.



```python
# Get unique values from the 'País' column
unique_countries = df['País'].unique()

# Print the unique values
print(unique_countries)
```

## Crear un mapeo de corrección

### Subtask:
Definir un diccionario o mapeo para corregir los valores inconsistentes (abreviaciones, errores de escritura, caracteres especiales).


**Reasoning**:
Define the country_mapping dictionary as specified in the instructions.



```python
country_mapping = {
    'AR': 'Argentina',
    'BO': 'Bolivia',
    'MÉ': 'México',
    'PE': 'Perú',
    'UR': 'Uruguay',
    'VE': 'Venezuela',
    'EC': 'Ecuador',
    'CO': 'Colombia',
    'CH': 'Chile',
    'ES': 'España',
    'Arg3ntin@': 'Argentina',
    'P3rú': 'Perú',
    'Urugu@y': 'Uruguay',
    'Esp@ñ@': 'España',
    'V3n3zu3l@': 'Venezuela',
    'Chil3': 'Chile',
    'Ecu@dor': 'Ecuador',
    'Bolivi@': 'Bolivia',
    'Colombi@': 'Colombia',
    'Uruguay Uruguay': 'Uruguay',
    'Bolivia Bolivia': 'Bolivia',
    'Perú Perú': 'Perú',
    'México México': 'México',
    'Venezuela Venezuela': 'Venezuela',
    'Chile Chile': 'Chile',
    'Argentina Argentina': 'Argentina',
    'España España': 'España',
    'Ecuador Ecuador': 'Ecuador',
    'Colombia Colombia': 'Colombia',
    'Madrid': 'España',
    'Bogotá': 'Colombia',
    'Ciudad de México': 'México',
    'no_dato': 'no_dato'
}
```

## Aplicar las correcciones

### Subtask:
Usar el mapeo para reemplazar los valores inconsistentes en la columna 'País'.


**Reasoning**:
Replace the inconsistent values in the 'País' column using the defined mapping and verify the changes.



```python
# Replace inconsistent values in the 'País' column using the mapping
df['País'] = df['País'].replace(country_mapping)

# Display the unique values in the 'País' column to verify the changes
print(df['País'].unique())
```

**Reasoning**:
The previous step did not completely replace the inconsistent country names; some entries with extra spaces or repetitions were not corrected. I will attempt to remove leading/trailing spaces and replace the remaining repetitive entries using the mapping again to ensure complete cleaning of the 'País' column.



```python
# Remove leading and trailing spaces from the 'País' column
df['País'] = df['País'].str.strip()

# Apply the country_mapping again to handle potential remaining inconsistencies
df['País'] = df['País'].replace(country_mapping)

# Display the unique values in the 'País' column to verify the changes
print(df['País'].unique())
```

## Verificar los cambios

### Subtask:
Mostrar los valores únicos de la columna 'País' nuevamente para confirmar que las inconsistencias han sido corregidas.


**Reasoning**:
Display the unique values of the 'País' column to confirm the cleaning process.



```python
# Get the unique values from the 'País' column
unique_countries_after_cleaning = df['País'].unique()

# Print the unique values
print(unique_countries_after_cleaning)
```

## Summary:

### Data Analysis Key Findings

*   The initial unique values in the 'País' column included abbreviations (e.g., 'AR', 'BO'), misspellings and special characters (e.g., 'Arg3ntin@', 'P3rú'), repeated country names (e.g., 'Uruguay Uruguay'), city names (e.g., 'Madrid', 'Bogotá'), and a placeholder 'no\_dato'.
*   A `country_mapping` dictionary was created to map these inconsistent values to their correct full names.
*   Applying the `replace()` method with the `country_mapping` initially corrected many inconsistencies but did not handle entries with leading/trailing spaces or repeated names within a single entry effectively.
*   After stripping leading and trailing spaces from the 'País' column and reapplying the mapping, the column was successfully cleaned.
*   The final unique values in the 'País' column are now a consistent list of country names ('España', 'Argentina', 'Colombia', 'Uruguay', 'Perú', 'Bolivia', 'México', 'Venezuela', 'Chile', 'Ecuador') and 'no\_dato'.

```python
# Convert the 'Fecha_Registro' column to datetime objects, coercing errors
df['Fecha_Registro'] = pd.to_datetime(df['Fecha_Registro'], errors='coerce')

# Format the 'Fecha_Registro' column to display only the date
df['Fecha_Registro'] = df['Fecha_Registro'].dt.strftime('%Y-%m-%d')

# Display the first few rows of the DataFrame to verify the changes
print(df.head(10))
```

```python
# Calculate the number of 'no_dato' values in the 'Comentarios' column
no_dato_count = (df['Comentarios'] == 'no_dato').sum()

# Calculate the total number of rows in the DataFrame
total_rows = len(df)

# Calculate the proportion of 'no_dato' values
no_dato_proportion = no_dato_count / total_rows

# Define a threshold (e.g., 50%)
threshold = 0.5

# Check if the proportion of 'no_dato' values is above the threshold
if no_dato_proportion > threshold:
    # Drop the 'Comentarios' column
    df = df.drop('Comentarios', axis=1)
    print("La columna 'Comentarios' ha sido eliminada.")
else:
    print("La columna 'Comentarios' no tiene una mayoría de valores 'no_dato' y no será eliminada.")

# Display the columns of the DataFrame to verify the change
print(df.columns)
```

```python
# Create a new numerical ID column
df['ID_Numerico_Encuesta'] = range(1, len(df) + 1)

# Drop the original 'ID_Encuesta' column if you no longer need it
df = df.drop('ID_Encuesta', axis=1)

# Display the first few rows with the new ID to verify
print(df.head())
```

```python
# Define the path for the new cleaned CSV file
output_path = '/content/dataset_limpio.csv'

# Save the cleaned DataFrame to a new CSV file
df.to_csv(output_path, index=False)

print(f"El dataset limpio ha sido guardado en: {output_path}")
```
