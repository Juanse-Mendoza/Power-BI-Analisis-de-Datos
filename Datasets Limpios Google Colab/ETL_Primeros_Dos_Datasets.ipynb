```python
import pandas as pd

# Define la ruta del archivo de Excel.
# Reemplaza 'nombre_de_tu_archivo.xlsx' con el nombre de tu archivo real.
ruta_archivo = '/content/Dataset con varios errores 2.xlsx'

# Carga el archivo de Excel en un DataFrame de pandas.
df = pd.read_excel(ruta_archivo)

# Muestra las primeras 5 filas del DataFrame para verificar que se cargó correctamente.
print(df.head())
```

```python

# Remplaza todos los valores nulos (NaN) en la columna 'Nombre_Usuario' por 'sin_nombre'
df['Nombre_Usuario'] = df['Nombre_Usuario'].fillna('sin_nombre')

# Muestra el DataFrame modificado para verificar el cambio
print(df.head(8))
```

```python
# Reemplaza los valores de texto por sus equivalentes numéricos
df['Edad'] = df['Edad'].replace({'treinta y siete': 37,
                                 'veinticinco': 25})

# Muestra el DataFrame modificado
print(df.head())
```

```python
# Reemplaza la cadena de texto 'null' por 'Sin_ciudad'
df['País'] = df['País'].replace('null', 'sin_ciudad')

# Reemplaza los valores nulos (NaN) por 'Sin_ciudad'
df['País'] = df['País'].fillna('sin_ciudad')

#Ahora se cambia el formato de escritura de los paises para que queden iguales los formatos
# Convierte todos los valores de la columna 'Pais' a formato título
df['País'] = df['País'].astype(str).str.strip().str.title()

# Muestra el DataFrame modificado
print(df.head(8))
```

```python
# Reemplaza la cadena de texto 'null' por 'Sin_ciudad'
df['Plan_Suscripcion'] = df['Plan_Suscripcion'].replace('null', 'sin_suscripcion')

# Reemplaza los valores nulos (NaN) por 'Sin_ciudad'
df['Plan_Suscripcion'] = df['Plan_Suscripcion'].fillna('sin_suscripcion')

#Ahora se cambia el formato de escritura de los paises para que queden iguales los formatos
# Convierte todos los valores de la columna 'Pais' a formato título
df['Plan_Suscripcion'] = df['Plan_Suscripcion'].astype(str).str.strip().str.title()

# Muestra el DataFrame modificado
print(df.head(10))
```

```python
# Convert the 'Minutos_Vistos_Mes' column to numeric, coercing errors
df['Minutos_Vistos_Mes'] = pd.to_numeric(df['Minutos_Vistos_Mes'], errors='coerce')

# Display the modified DataFrame to verify the change
print(df.head(10))
```

```python
# Reemplaza los valores de texto por sus equivalentes numéricos
df['Calificacion_Servicio'] = df['Calificacion_Servicio'].replace({'Seis': 6,
                                 })

# Muestra el DataFrame modificado
print(df.head(8))
```

```python
# Remplaza todos los valores nulos (NaN) en la columna 'Producto_Favorito' por 'sin_producto'
df['Genero_Favorito'] = df['Genero_Favorito'].fillna('sin_genero')

# Muestra el DataFrame modificado para verificar el cambio
print(df.head())
```

```python
# Reemplaza la cadena de texto 'null' por 'Sin_comentarios'
df['Comentarios'] = df['Comentarios'].replace('null', 'sin_comentarios')

# Reemplaza los valores nulos (NaN) por 'Sin_ciudad'
df['Comentarios'] = df['Comentarios'].fillna('sin_comentarios')

# Muestra el DataFrame modificado
print(df.head(10))
```

```python
# Define la ruta y el nombre del nuevo archivo de Excel
ruta_archivo_final = 'mi_dataframe_limpio_2.xlsx'

# Exporta el DataFrame a un archivo de Excel
df.to_excel(ruta_archivo_final, index=False)

print(f"DataFrame exportado exitosamente a '{ruta_archivo_final}'")
```
